# Análisis de Perfumes en eBay
Este notebook realiza comparaciones visuales entre perfumes disponibles en eBay.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
file_path = "Salary.csv"  # Asegúrate de ubicar el archivo correctamente
df = pd.read_csv(file_path)

# Configuración de estilo de gráficos
sns.set(style="whitegrid")

# Mostrar las primeras filas para entender la estructura
df.head()
```

# 1. Comparación de salario entre géneros

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Gender", y="Salary", data=df, palette="coolwarm")
plt.title("Salario Promedio por Género")
plt.show()
```

# 2. Comparación de salario por nivel educativo

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Education Level", y="Salary", data=df, palette="viridis")
plt.title("Salario Promedio por Nivel Educativo")
plt.show()
```

# 3. Comparación de salario entre seniors y no seniors

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Senior", y="Salary", data=df, palette="pastel")
plt.xticks(ticks=[0, 1], labels=["No Senior", "Senior"])
plt.title("Comparación de Salario entre Seniors y No Seniors")
plt.show()
```

# 4. Comparación de salario por país

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Country", y="Salary", data=df, palette="magma")
plt.title("Salario Promedio por País")
plt.xticks(rotation=45)
plt.show()
```
